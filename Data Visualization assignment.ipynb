```python
import numpy as np,matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
# loading the dataset
df=datasets.load_iris()
```

```python
df
```

```python
#PCA is use to reduce the number of dimensions in the array as above we can see we have 4 dimension in the array so we have to convert it into 3 dimension to plot a 3d plot
x_red=PCA(n_components=3).fit_transform(df['data']) #dimensionality reduction
```

```python
#plotting figure
fig=plt.figure(figsize=(22,109))
ax=Axes3D(fig,elev=50,azim=140,auto_add_to_figure=False)
fig.add_axes(ax)
ax.scatter(x_red[:,0],x_red[:,1],x_red[:,2],s=200,marker='*',c=df.target,cmap=plt.cm.Dark2_r)
ax.set_title("3d Iris Plot",fontsize=15) 
ax.set_xlabel("X axis",fontsize=15)
ax.set_ylabel("Y axis",fontsize=15)
ax.set_zlabel("Z axis",fontsize=15);
```
